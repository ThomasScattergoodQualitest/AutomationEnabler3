```python
import nltk
import csv
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words("english"))
```

```python
JL = open("JlAllReviews.txt")
```

```python
JL_text = JL.read()
```

```python
JL_text
```

```python
JL_token = nltk.word_tokenize(JL_text)
```

```python
nltk.FreqDist(JL_token)
```

```python
for word in JL_token:
        print (word)
```

```python
norm = [word.lower()for word in JL_token if word.isalpha()]
```

```python
norm[:10]
```

```python
filtered_words = [word for word in norm if word not in stop_words]
```

```python
filtered_words
```

```python
fw_fd = nltk.FreqDist(filtered_words) 
```

```python
fw_fd
```

```python
fw_fd.plot(25, cumulative=False) 
```
