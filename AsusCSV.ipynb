```python
data = pd.read_csv('AsusReviewdata.csv')
print ("Original 'AsusReviewdata.csv' CSV Data: \n")
data.columns=['Index','Date','Review','Rating']
del data['Index']
data
```

```python
comments = []
with open ("AsusReviewdata.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
        print (row)
```

```python
del comments [0]
```

```python
review = comments[0][2].lower()
print (review)
blob = TextBlob(review)
sentiment = blob.sentiment
print (sentiment)
```

```python
polarities = []
subjectivities = []
for doc in comments: 
    blob = TextBlob(doc[2])
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    polarities.append(polarity)
    subjectivities.append(subjectivity)
```

```python
plt.scatter (polarities, subjectivities)
plt.xlabel ("polarities")
plt.ylabel ("subjectivities")
plt.title ("subjectivities as a function of polarities of Asus reviews")
plt.show
```

```python
data ["Subjectivity"] = subjectivities 
data ["Polarity"] = polarities
full_data = data
full_data
```

```python
pol_mean = full_data['Polarity'].mean()
sub_mean = full_data['Subjectivity'].mean()
print ("average polarity is:")
print(pol_mean)
print ("average subjectvitiy is:")
print (sub_mean)
```

```python
full_data.to_csv('AsusFinal.csv', index=False)
```
