```python
import nltk
import csv
from nltk.corpus import stopwords
import string
import spacy
import numpy as np
import re
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
comments = []
with open ("reviews_ioCombined.csv", "r", encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
        print (row)
```

```python
comments[1][1].lower()
```

```python
blob = TextBlob(comments[420][1])
sentiment = blob.sentiment.polarity
print (sentiment)
```

```python
polarities = []
subjectivities = []
for doc in comments: 
    blob = TextBlob(doc[1])
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    polarities.append(polarity)
    subjectivities.append(subjectivity)
```

```python
plt.scatter (polarities, subjectivities)
plt.xlabel ("polarities")
plt.ylabel ("subjectivities")
plt.title ("subjectivities as a function of polarities of reviews.io reviews")
plt.show
```
