```python
import nltk
import csv
from nltk.corpus import stopwords
import string
import spacy
import matplotlib.pyplot as plt
import numpy as np
import re
import pandas as pd
from textblob import TextBlob
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
comments = []
with open ("KFCReviews.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        comments.append(row)
        print (row)
```

```python
comments[2][0].lower()
```

```python
blob = TextBlob(comments[2][0])
sentiment = blob.sentiment.polarity
print (sentiment)
```

```python
polarities = []
subjectivities = []
for doc in comments: 
    blob = TextBlob(doc[0])
    polarity = blob.sentiment.polarity
    subjectivity = blob.sentiment.subjectivity
    polarities.append(polarity)
    subjectivities.append(subjectivity)
```

```python
plt.scatter (polarities, subjectivities)
plt.xlabel ("polarities")
plt.ylabel ("subjectivities")
plt.title ("subjectivities as a function of polarities of KFC reviews")
plt.show
```

```python
for review in comments:
    print ("\n")
    print (np.average(sentiment(str(review[0]))))
    print(review[0])
```
